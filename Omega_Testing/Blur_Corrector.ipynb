```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
img = cv2.imread("test_img.png")
plt.imshow(img)
```

```python
kernel = np.array([[-2,-2,-2], [-2,18,-2], [-2,-2,-2]])
img = cv2.filter2D(img, -1, kernel)
plt.imshow(img)
```

```python
cv2.imwrite("Corrected.jpg",img)
```

```python
img_2 = cv2.imread("X-Ray-Images/Motion_Blur.jpg")
plt.imshow(img_2)
```

```python
close_kernel = np.ones((7,7))

canny = cv2.Canny(img_2,200,250)
closing = cv2.morphologyEx(canny,cv2.MORPH_CLOSE,close_kernel)
cv2.imwrite("Canny_Corrected.jpg",closing)
plt.imshow(closing)
```

```python
img_2 = cv2.imread("X-Ray-Images/Blurred.jpg")
print(img_2.shape)
b25 = img_2[int(img_2.shape[0]*0.75):,:]
plt.imshow(b25)
```

```python
var = cv2.Laplacian(b25, cv2.CV_64F).var()
print(var)
```

```python
img_gray = cv2.cvtColor(img_2,cv2.COLOR_BGR2GRAY)
ret,thresh = cv2.threshold(img_gray,140,255,cv2.THRESH_BINARY_INV)
cv2.imwrite("Canny_Corrected.jpg",thresh)
plt.imshow(thresh)
```
