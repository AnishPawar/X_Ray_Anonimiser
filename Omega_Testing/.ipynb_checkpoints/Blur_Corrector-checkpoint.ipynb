```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
img = cv2.imread("test_img.png")
plt.imshow(img)
```

```python
kernel = np.array([[-2,-2,-2], [-2,18,-2], [-2,-2,-2]])
img = cv2.filter2D(img, -1, kernel)
plt.imshow(img)
```

```python
cv2.imwrite("Corrected.jpg",img)
```
